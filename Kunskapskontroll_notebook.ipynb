```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("Too low balance")
        else:
            self.balance -= amount
            print(f"Withdrawn: ${amount}. New balance: ${self.balance}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: ${amount}. New balance: ${self.balance}")
```

```python
account1 = BankAccount("Damien", 100)

account1.deposit(50)

print(account1.account_holder)
```

```python
account1 = BankAccount("Damien", 150)

account1.withdraw(100)

print(account1.account_holder)
```

```python
account1 = BankAccount("Damien", 150)

account1.withdraw(200)

print(account1.account_holder)
```

## Part 2

 In the diamond industry, quality grades like cut, color, and clarity heavily influence pricing — but are premium-grade diamonds truly worth the extra cost? This analysis explores how these grading factors affect diamond prices, using a dataset of over 50,000 diamonds. By comparing the price distributions across different quality levels, we aim to uncover whether higher grades command significantly higher prices, and if the price differences align with perceptible differences in quality. This insight can help guide both consumers and us sellers in making informed decisions about diamond value.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("diamonds.csv")

df_clean = df.dropna(subset=["price", "cut", "color", "clarity"])
```

```python
avg_price_cut = df_clean.groupby("cut")["price"].mean().sort_values(ascending=False)
avg_price_cut.plot(kind="bar", figsize=(8, 5), title="Average Diamond Price by Cut", color="skyblue")
plt.xlabel("Cut")
plt.ylabel("Average Price ($)")
plt.tight_layout()
plt.show()
```

This bar chart shows the average price of diamonds by cut quality level.
Surprisingly, Fair cuts have almost as high of an average price as premium cuts do. Ideal cuts, often considered the highest quality, actually have the lowest average price. Based on the data, cut quality does not directly correlate with price. Premium and fair cuts likely have average carat sizes, driving up their prices more than ideal cuts.

```python
avg_price_color = df_clean.groupby("color")["price"].mean()
avg_price_color.plot(kind="line", marker='o', figsize=(8, 5), title="Average Diamond Price by Color Grade",
                     color="green")
plt.xlabel("Color Grade")
plt.ylabel("Average Price ($)")
plt.tight_layout()
plt.show()
```

The diamond price based on the quality of the color seems to be reverted, the lower quality coloring like J seems to have the highest average price by far.
this might relate to the actual carat size of the diamond.

```python
df_clean.groupby("clarity")["price"].plot(kind="hist", bins=30, alpha=0.5, legend=True, figsize=(10, 6))
plt.title("Price Distribution Across Clarity Grades")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

There seems to be quite a significant difference in frequency of distributed diamonds within the medium to lower clarity. This might indicate to a seller that investing in the medium clarity diamonds is a more optimal financial decision.

This overall analysis compares diamond prices across three key quality metrics: cut, color, clarity. While conventional wisdom suggests that higher grades in these categories command significantly higher prices, the data tells a more nuanced story. Color and clarity demonstrate subtle and overlapping price effects, especially in mid-tier grades. Customers usually want and can find good value in mid-range quality grades without paying extreme premiums. For businesses. this insight supports tiered pricing strategies and value-based product positioning.
